```python
## Combine the data from before and after motif in one file to plot the figures for long disorder prediction

import pandas as pd
import numpy as np

def merge_score_data(data_before, data_after):
    ELM_mean=data_before.elm_mean
    data_before=data_before.iloc[: , -1].str.split(',', expand=True).replace( ["\\[", "\\]", "'"], "", regex=True)
    data_after=data_after.iloc[: , -1].str.split(',', expand=True).replace( ["\\[", "\\]", "'"], "", regex=True)
    merged_data=[data_before,ELM_mean,data_after]
    merged_df=pd.concat(merged_data,axis=1, ignore_index=True)
    merged_df.columns=list(map(str,(list(range(-100,101)))))
    return merged_df

#Eukaryotes Disorder profile for SLiMs and 100 residues before and after
eukaryotes_diso_before=pd.read_csv('eukaryotes/eukaryotes_long_disorder_before_motif_data_100.csv')
eukaryotes_diso_after=pd.read_csv('eukaryotes/eukaryotes_long_disorder_after_motif_data_100.csv')

merged_eukaryotes_long_diso_data=merge_score_data(eukaryotes_diso_before,eukaryotes_diso_after)


#Bacteria Disorder profile for SLiMs and 100 residues before and after

bacteria_diso_before=pd.read_csv('bacteria/bacteria_long_disorder_before_motif_data_100.csv')
bacteria_diso_after=pd.read_csv('bacteria/bacteria_long_disorder_after_motif_data_100.csv')

merged_bacteria_long_diso_data=merge_score_data(bacteria_diso_before,bacteria_diso_after)


#Viruses Disorder profile for SLiMs and 100 residues before and after


viruses_diso_before=pd.read_csv('viruses/viruses_long_disorder_before_motif_data_100.csv')
viruses_diso_after=pd.read_csv('viruses/viruses_long_disorder_after_motif_data_100.csv')

merged_viruses_long_diso_data=merge_score_data(viruses_diso_before,viruses_diso_after)


#save the generated files
merged_eukaryotes_long_diso_data.to_csv("merged_eukaryotes_long_diso_data.csv",encoding='utf-8', index=False, header= True )
merged_bacteria_long_diso_data.to_csv("merged_bacteria_long_diso_data.csv",encoding='utf-8', index=False, header= True )
merged_viruses_long_diso_data.to_csv("merged_viruses_long_diso_data.csv",encoding='utf-8', index=False, header= True )
```

## Plot the long disorder profile data for the SLiMs and the 100 residues flanking regions

```python
import numpy as np
import scipy.stats as st
import math
import seaborn as sns
import matplotlib.pyplot as plt

def get_mean_and_CI(df):
    def skip_none(a):
        without_none=[]
        for value in a:
            if value is '' or value is 'nan':
                pass
            else:
                without_none.append(float(value))
        return without_none
    col_means=[]
    LI_UI=[]
    for (col_name, col_data) in df.iteritems():
        df_len=len(col_data)
        col_data=skip_none(col_data)
        col_mean=float(np.nanmean(col_data))
        col_std=np.nanstd(col_data, ddof=1)
        sigma= col_std/math.sqrt(df_len)
        LI_UI_values=st.t.interval(alpha=0.95, df=df_len-1, loc=col_mean, scale=sigma)
        col_means.append(col_mean)
        LI_UI.append(LI_UI_values)
    return col_means, LI_UI

euk_col_means, euk_LI_UI=get_mean_and_CI(merged_eukaryotes_long_diso_data)
euk_LI=[x[0] for x in euk_LI_UI]
euk_UI=[x[1] for x in euk_LI_UI]
                        ##############################################################
bac_col_means, bac_LI_UI=get_mean_and_CI(merged_bacteria_long_diso_data)
bac_LI=[x[0] for x in bac_LI_UI]
bac_UI=[x[1] for x in bac_LI_UI]
                    ###################################################################
virus_col_means, virus_LI_UI = get_mean_and_CI(merged_viruses_long_diso_data)
virus_LI=[x[0] for x in virus_LI_UI]
virus_UI=[x[1] for x in virus_LI_UI]                   


                       ##################### Plot the data ##########################
x=list(range(0,201,1))

fig, ax = plt.subplots(sharey=True,figsize=[20, 10])
sns.set(font_scale = 3)
sns.set_style('ticks') 

plt.plot(euk_col_means, color='black')
plt.plot(euk_LI, color='grey')
plt.plot(euk_UI, color='grey')
plt.fill_between(x, y1= euk_UI, y2= euk_LI, color='lightgrey', alpha =0.4)

plt.plot(bac_col_means, color='darkgreen')
plt.plot(bac_LI, color='lightgreen')
plt.plot(bac_UI, color='lightgreen')
plt.fill_between(x, y1= bac_UI, y2= bac_LI, color='lightgreen', alpha =0.4)

plt.plot(virus_col_means, color='darkblue')
plt.plot(x,virus_UI, color='lightblue')
plt.plot(x,virus_LI, color='lightblue')
plt.fill_between(x, y1= virus_UI, y2= virus_LI, color='lightblue', alpha =0.4)


plt.axvline(100, color='black', linewidth=2, alpha=0.2)


plt.xticks(list(range(0,201,20)), ['-100','-80', '-60', '-40', '-20','0', '20', '40','60', '80', '100'], rotation = 90, fontsize=28)
plt.yticks( fontsize=28)
plt.ylim(0,1)
ax.set_xlabel("Residues",fontsize=36, fontweight='bold')
ax.set_ylabel("Long IUPRED2A disorder score",fontsize=36, fontweight='bold')
ax = plt.gca()
ax.tick_params(width=3)
ax.spines['bottom'].set_linewidth(4)
ax.spines['top'].set_linewidth(4)
ax.spines['left'].set_linewidth(4)
ax.spines['right'].set_linewidth(4)

ax.get_figure().savefig('SLiMs_long_disorder_profile.png', dpi=600, bbox_inches='tight')
```

```python
import numpy as np
import scipy.stats as st
import math
import seaborn as sns
import matplotlib.pyplot as plt

def get_mean_and_CI(df):
    col_means=[]
    LI_UI=[]
    for (col_name, col_data) in df.iteritems():
        df_len=len(col_data)
        col_data=[float(x) for x in col_data]
        col_mean=float(np.nanmean(col_data))
        col_std=np.nanstd(col_data, ddof=1)
        sigma= col_std/math.sqrt(df_len)
        LI_UI_values=st.t.interval(alpha=0.95, df=df_len-1, loc=col_mean, scale=sigma)
        col_means.append(col_mean)
        LI_UI.append(LI_UI_values)
    return col_means, LI_UI

euk_col_means, euk_LI_UI=get_mean_and_CI(merged_eukaryotes_cc_data)
euk_LI=[x[0] for x in euk_LI_UI]
euk_UI=[x[1] for x in euk_LI_UI]
                        ##############################################################
bac_col_means, bac_LI_UI=get_mean_and_CI(merged_bacteria_cc_data)
bac_LI=[x[0] for x in bac_LI_UI]
bac_UI=[x[1] for x in bac_LI_UI]
                    ###################################################################
virus_col_means, virus_LI_UI = get_mean_and_CI(merged_viruses_cc_data)
virus_LI=[x[0] for x in virus_LI_UI]
virus_UI=[x[1] for x in virus_LI_UI]                   


                       ##################### Plot the data ##########################
x=list(range(0,201,1))

fig, ax = plt.subplots(sharey=True,figsize=[20, 10])
sns.set(font_scale = 3)
sns.set_style('ticks') 

plt.plot(euk_col_means, color='black')
plt.plot(euk_LI, color='grey')
plt.plot(euk_UI, color='grey')
plt.fill_between(x, y1= euk_UI, y2= euk_LI, color='lightgrey', alpha =0.4)

plt.plot(bac_col_means, color='darkgreen')
plt.plot(bac_LI, color='lightgreen')
plt.plot(bac_UI, color='lightgreen')
plt.fill_between(x, y1= bac_UI, y2= bac_LI, color='lightgreen', alpha =0.4)

plt.plot(virus_col_means, color='darkblue')
plt.plot(x,virus_UI, color='lightblue')
plt.plot(x,virus_LI, color='lightblue')
plt.fill_between(x, y1= virus_UI, y2= virus_LI, color='lightblue', alpha =0.4)


plt.axvline(100, color='black', linewidth=2, alpha=0.2)


plt.xticks(list(range(0,801,100)), ['-400','-300', '-200', '-100', '0', '100', '200', '300', '400'], rotation = 90, fontsize=28)
plt.ylim(0,1)
ax.set_xlabel("Residues",fontsize=36, fontweight='bold')
ax.set_ylabel("Coil confidence score",fontsize=36, fontweight='bold')
ax = plt.gca()
ax.tick_params(width=3)
ax.spines['bottom'].set_linewidth(4)
ax.spines['top'].set_linewidth(4)
ax.spines['left'].set_linewidth(4)
ax.spines['right'].set_linewidth(4)

ax.get_figure().savefig('SLiMs_coil_conf_profile.png', dpi=600, bbox_inches='tight')
```
